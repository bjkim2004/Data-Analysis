## 2. Stateless (Object-based) 접근법

### 2-1. Method 1: 도화지(Figure) 객체 생성

```python
fig = plt.figure(
    figsize=(8, 6),
    facecolor='#c1f1f1'
)
plt.show()
```

### 2-2. Method 1: Axes 좌표계 장착

```python
fig = plt.figure(figsize=(8, 6), facecolor='#c1f1f1')

rect = (0.1, 0.1, 0.8, 0.8)
ax = fig.add_axes(
    rect,
    facecolor='#e1e1e1'
)
plt.show()
```

### 2-3. Method 1: 데이터 매핑 및 출력

```python
fig = plt.figure(figsize=(8, 6), facecolor='#c1f1f1')
rect = (0.1, 0.1, 0.8, 0.8)
ax = fig.add_axes(rect, facecolor='#e1e1e1')

ax.scatter(iris['sepal_length'], iris['sepal_width'])

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_title('Completed Scatter Plot')

plt.show()
```

### 2-4. Method 2: Figure & Ax 동시 생성 (plt.subplots)

```python
fig, ax = plt.subplots(
    figsize=(8, 6),
    facecolor='#c1f1f1'
)

ax.scatter(iris['sepal_length'], iris['sepal_width'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_title('Scatter Plot with plt.subplots()')

plt.show()
```

### 2-5. Method 3: 다중 차트 그리드 생성

```python
fig, axes = plt.subplots(
    2, 2,
    figsize=(10, 8),
    facecolor='#c1f1f1'
)

axes[0, 0].scatter(iris['sepal_length'], iris['sepal_width'])
axes[0, 0].set_title('Sepal: Length vs Width')
axes[0, 0].set_xlabel('sepal_length')

axes[0, 1].scatter(iris['petal_length'], iris['petal_width'])
axes[0, 1].set_title('Petal: Length vs Width')
axes[0, 1].set_xlabel('petal_length')

axes[1, 0].bar(iris['sepal_length'], iris['sepal_width'])
axes[1, 0].set_xlabel('sepal_length')
axes[1, 0].set_ylabel('sepal_width')

axes[1, 1].bar(iris['petal_length'], iris['petal_width'])
axes[1, 1].set_xlabel('petal_length')
axes[1, 1].set_ylabel('petal_width')

plt.tight_layout()
plt.show()
```

## 3. Stateful (State-based) 접근법

### 3-1. 포커스(Focus) 활성화 및 플로팅

```python
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.scatter(iris['sepal_length'], iris['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('sepal_length vs sepal_width scatter')

plt.subplot(2, 2, 2)
plt.scatter(iris['petal_length'], iris['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('petal_length vs petal_width scatter')

plt.subplot(2, 2, 3)
plt.bar(iris['sepal_length'], iris['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('sepal_length vs sepal_width bar')

plt.subplot(2, 2, 4)
plt.bar(iris['petal_length'], iris['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
plt.title('Petal_length vs petal_width bar')

plt.suptitle('matplotlib statful style')
plt.tight_layout()
plt.show()
```

## 4. 라인 차트 (Line Chart)

### 4-1. 기초 라인 차트 렌더링

```python
plt.plot([0, 1, 2, 3], [3, 1, 5, 2])
plt.show()
```

### 4-2. 연속형 데이터 생성 (np.linspace + np.sin)

```python
x = np.linspace(0, 2*np.pi, 50)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.show()
```

### 4-3. 다중 라인 중복 렌더링

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.title('Sin & Cos Overlay')
plt.show()
```

### 4-4. 캔버스 분할 (Subplots)

```python
x = np.linspace(0, 2*np.pi, 100)

plt.subplot(121)
plt.plot(x, np.sin(x))
plt.title('Sin')

plt.subplot(122)
plt.plot(x, np.cos(x))
plt.title('Cos')

plt.tight_layout()
plt.show()
```

### 4-5. 축 범위 및 타이틀 제어

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)

plt.title('Set Limitation of Ax')
plt.xlim(0, 10)
plt.ylim(-3, 3)

plt.show()
```

### 4-6. 차트 스타일 지정 (Format Strings)

```python
x = np.linspace(0, 2*np.pi, 30)

plt.plot(x, np.sin(x), 'bs', label='sin')
plt.plot(x, np.cos(x)/2, 'r--^', label='cos/2')

plt.legend()
plt.title('Format Strings Example')
plt.show()
```

### 4-7. 이중 축 (Dual Y-Axis) 렌더링

```python
x = np.linspace(0, 2*np.pi, 100)

ax1 = plt.gca()
ax2 = plt.gca().twinx()

ax1.plot(x, np.sin(x), label='sin')
ax2.plot(x, np.cos(x)/2, color='r', label='cos/2')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Dual Y-Axis Chart')
plt.show()
```

## 8. 다변량 산점도 (Multivariate Scatter Plot)

### Step 1. 데이터셋 로드

```python
tips = sns.load_dataset('tips')
print(tips.head())
```

### Step 2. Figure 및 Subplot 설정

```python
scatter_plot = plt.figure()
scatter_plot.set_figwidth(10)
scatter_plot.set_figheight(7)

axes1 = scatter_plot.add_subplot(1, 1, 1)
```

### Step 3. 이변량 기본 산점도 그리기

```python
scatter_plot = plt.figure()
scatter_plot.set_figwidth(10)
scatter_plot.set_figheight(7)
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(
    tips['total_bill'],
    tips['tip']
)

axes1.set_title('Total Bill vs Tip')
axes1.set_xlabel('Total_bill')
axes1.set_ylabel('tip')

plt.show()
```

### Step 4. 세 번째 변수 추가 (크기 조절)

```python
scatter_plot = plt.figure()
scatter_plot.set_figwidth(10)
scatter_plot.set_figheight(7)
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(
    tips['total_bill'],
    tips['tip'],
    s=tips['size'] * 20
)

axes1.set_title('Total Bill vs Tip (size = party size)')
axes1.set_xlabel('Total_bill')
axes1.set_ylabel('tip')

plt.show()
```

### Step 5. 네 번째 변수 추가 (색상 구분)

```python
tips['sex_color'] = tips['sex'].map({'Female': 0, 'Male': 1})

scatter_plot = plt.figure()
scatter_plot.set_figwidth(10)
scatter_plot.set_figheight(7)
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(
    tips['total_bill'],
    tips['tip'],
    s=tips['size'] * 20,
    c=tips['sex_color'],
    cmap='cool'
)

axes1.set_title('Total bill vs tip colored by sex_color, size')
axes1.set_xlabel('Total_bill')
axes1.set_ylabel('tip')

plt.show()
```

### Step 6. 투명도 조정 (alpha)

```python
scatter_plot = plt.figure()
scatter_plot.set_figwidth(10)
scatter_plot.set_figheight(7)
axes1 = scatter_plot.add_subplot(1, 1, 1)

axes1.scatter(
    tips['total_bill'],
    tips['tip'],
    s=tips['size'] * 20,
    c=tips['sex_color'],
    cmap='cool',
    alpha=0.5
)

axes1.set_title('Total bill vs tip (with alpha=0.5)')
axes1.set_xlabel('Total_bill')
axes1.set_ylabel('tip')

plt.show()
```

## 9. 상관관계 히트맵 (Correlation Heatmap)

### Step 1. 상관계수 행렬 생성

```python
corr = tips[['total_bill', 'tip', 'size']].corr()
display(corr)
```

### Step 2. pcolor 히트맵 기본 격자 생성

```python
plt.figure(figsize=(8, 6))
plt.pcolor(corr)
plt.title('Basic pcolor Heatmap')
plt.show()
```

### Step 3. 축 틱(Ticks) 및 레이블 포지셔닝

```python
plt.figure(figsize=(8, 6))
plt.pcolor(corr)

ncolumns = corr.columns
plt.xticks(np.arange(0.5, len(ncolumns), 1), ncolumns)
plt.yticks(np.arange(0.5, len(ncolumns), 1), ncolumns)

plt.title('Heatmap with Labels')
plt.show()
```

### Step 4. 색상 범례 (Colorbar) 추가

```python
plt.figure(figsize=(8, 6))
plt.pcolor(corr)

ncolumns = corr.columns
plt.xticks(np.arange(0.5, len(ncolumns), 1), ncolumns)
plt.yticks(np.arange(0.5, len(ncolumns), 1), ncolumns)

plt.colorbar()
plt.title('Correlation Heatmap with Colorbar')
plt.show()
```

## 10. 종합 실습

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Iris Dataset Visualization Dashboard', fontsize=16, fontweight='bold')

species_map = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
colors = iris['species'].map(species_map)
axes[0, 0].scatter(iris['sepal_length'], iris['sepal_width'], c=colors, cmap='viridis', alpha=0.7)
axes[0, 0].set_xlabel('Sepal Length')
axes[0, 0].set_ylabel('Sepal Width')
axes[0, 0].set_title('Sepal: Length vs Width by Species')

axes[0, 1].hist(iris['petal_length'], bins=20, edgecolor='black', color='skyblue')
axes[0, 1].set_xlabel('Petal Length')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Distribution of Petal Length')

setosa = iris[iris['species'] == 'setosa']['sepal_length']
versicolor = iris[iris['species'] == 'versicolor']['sepal_length']
virginica = iris[iris['species'] == 'virginica']['sepal_length']
axes[1, 0].boxplot([setosa, versicolor, virginica], labels=['Setosa', 'Versicolor', 'Virginica'])
axes[1, 0].set_ylabel('Sepal Length')
axes[1, 0].set_title('Sepal Length by Species')

species_mean = iris.groupby('species')[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].mean()
for col in species_mean.columns:
    axes[1, 1].plot(species_mean.index, species_mean[col], marker='o', label=col)
axes[1, 1].legend(fontsize=8)
axes[1, 1].set_ylabel('Mean Value')
axes[1, 1].set_title('Mean Features by Species')
axes[1, 1].tick_params(axis='x', rotation=15)

plt.tight_layout()
plt.show()
```

```python
iris_corr = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()

plt.figure(figsize=(8, 6))
plt.pcolor(iris_corr, cmap='coolwarm')

ncolumns = iris_corr.columns
plt.xticks(np.arange(0.5, len(ncolumns), 1), ncolumns, rotation=30)
plt.yticks(np.arange(0.5, len(ncolumns), 1), ncolumns)
plt.colorbar()
plt.title('Iris Feature Correlation Heatmap')

plt.tight_layout()
plt.show()
```
