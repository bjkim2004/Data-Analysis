## Part 2. 시각화 유형별 예제

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1) 시간 (Time) - 선 그래프
months = ['1월', '2월', '3월', '4월', '5월', '6월']
sales = [120, 135, 148, 160, 175, 190]
axes[0].plot(months, sales, marker='o', color='#1f77b4', linewidth=2)
axes[0].set_title('시간(Time): 월별 매출 추이', fontweight='bold')
axes[0].set_ylabel('매출(백만원)')
axes[0].grid(True, alpha=0.3)

# 2) 분포 (Distribution) - 파이 차트
labels = ['제품A', '제품B', '제품C', '제품D']
sizes = [35, 25, 25, 15]
colors = ['#2196F3', '#4CAF50', '#FF9800', '#F44336']
axes[1].pie(sizes, labels=labels, colors=colors, autopct='%1.0f%%', startangle=90)
axes[1].set_title('분포(Distribution): 제품별 비율', fontweight='bold')

# 3) 관계 (Relationship) - 산점도
np.random.seed(42)
x = np.random.randn(50)
y = x * 0.8 + np.random.randn(50) * 0.5
axes[2].scatter(x, y, alpha=0.7, color='#9C27B0', edgecolors='white')
axes[2].set_title('관계(Relationship): 변수 간 상관', fontweight='bold')
axes[2].set_xlabel('변수 X')
axes[2].set_ylabel('변수 Y')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Part 5. 차트 구성 요소 조작

```python
fig, ax = plt.subplots(figsize=(8, 5))

x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)

ax.plot(x, y1, 'o-', label='sin(x)', color='#2196F3', markersize=6, linewidth=2)
ax.plot(x, y2, 's--', label='cos(x)', color='#FF5722', markersize=5, linewidth=2)

ax.set_title('Matplotlib 차트 구성 요소 실습', fontsize=14, fontweight='bold')
ax.set_xlabel('X 값', fontsize=12)
ax.set_ylabel('Y 값', fontsize=12)
ax.grid(True, alpha=0.3, linestyle='--')
ax.legend(fontsize=11, loc='upper right')
ax.set_xlim(0, 10)
ax.set_ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()
```

## Part 6. Matplotlib 코딩 패러다임: OO vs Pyplot

### 6.1 OO (Object-Oriented) Style

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.set_title("OO Style: Simple Plot")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()
```

### 6.2 Pyplot Style

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.title("Pyplot Style: Simple Plot")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 9.3 데이터셋 그룹화 및 기술 통계량 산출

```python
ans = anscombe.groupby(anscombe['dataset'])
ans.describe()
```

```python
for name, group in ans:
    print(f"--- Dataset {name} ---")
    print(f"  x 평균: {group['x'].mean():.2f}  |  y 평균: {group['y'].mean():.4f}")
    print(f"  x 표준편차: {group['x'].std():.4f}  |  y 표준편차: {group['y'].std():.4f}")
    print(f"  상관계수: {group['x'].corr(group['y']):.4f}")
    print()
```

### 9.5 각 구역에 데이터 매핑

```python
fig = plt.figure(figsize=(10, 8))

axes1 = fig.add_subplot(2, 2, 1)
axes2 = fig.add_subplot(2, 2, 2)
axes3 = fig.add_subplot(2, 2, 3)
axes4 = fig.add_subplot(2, 2, 4)

ds1 = anscombe[anscombe['dataset'] == 'I']
ds2 = anscombe[anscombe['dataset'] == 'II']
ds3 = anscombe[anscombe['dataset'] == 'III']
ds4 = anscombe[anscombe['dataset'] == 'IV']

axes1.plot(ds1['x'], ds1['y'], 'o', color='#2196F3', markersize=8)
axes2.plot(ds2['x'], ds2['y'], 'o', color='#4CAF50', markersize=8)
axes3.plot(ds3['x'], ds3['y'], 'o', color='#FF9800', markersize=8)
axes4.plot(ds4['x'], ds4['y'], 'o', color='#F44336', markersize=8)

axes1.set_title('DS1')
axes2.set_title('DS2')
axes3.set_title('DS3')
axes4.set_title('DS4')

fig.suptitle('Anscombe Data', fontsize=16, fontweight='bold')
fig.tight_layout()
plt.show()
```

### 9.6 회귀선 추가 (Seaborn lmplot)

```python
g = sns.lmplot(
    data=anscombe,
    x='x', y='y',
    col='dataset',
    col_wrap=2,
    ci=None,
    height=4,
    aspect=1.2,
    scatter_kws={'s': 60, 'alpha': 0.8},
    line_kws={'color': 'red', 'linewidth': 2}
)
g.figure.suptitle('Anscombe Quartet + 회귀선', fontsize=14, fontweight='bold', y=1.02)
plt.show()
```

### 9.7 tight_layout() 전후 비교

```python
# Before
fig, axes_b = plt.subplots(2, 2, figsize=(6, 5))
for i, (name, group) in enumerate(ans):
    ax = axes_b[i // 2, i % 2]
    ax.plot(group['x'], group['y'], 'o')
    ax.set_title(f'DS{name}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
fig.suptitle('Before: tight_layout() 적용 전', fontweight='bold')
plt.show()

# After
fig, axes_a = plt.subplots(2, 2, figsize=(6, 5))
for i, (name, group) in enumerate(ans):
    ax = axes_a[i // 2, i % 2]
    ax.plot(group['x'], group['y'], 'o')
    ax.set_title(f'DS{name}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
fig.suptitle('After: tight_layout() 적용 후', fontweight='bold')
fig.tight_layout()
plt.show()
```

## Part 10. 종합 실습 과제

### 과제 1: penguins 데이터셋 시각화

```python
penguins = sns.load_dataset('penguins').dropna()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for species in penguins['species'].unique():
    subset = penguins[penguins['species'] == species]
    axes[0, 0].scatter(subset['bill_length_mm'], subset['bill_depth_mm'],
                        label=species, alpha=0.7, s=40)
axes[0, 0].set_title('산점도: 부리 길이 vs 깊이')
axes[0, 0].set_xlabel('부리 길이 (mm)')
axes[0, 0].set_ylabel('부리 깊이 (mm)')
axes[0, 0].legend()

axes[0, 1].hist(penguins['body_mass_g'], bins=20, color='#42A5F5', edgecolor='white')
axes[0, 1].set_title('히스토그램: 체중 분포')
axes[0, 1].set_xlabel('체중 (g)')
axes[0, 1].set_ylabel('빈도')

penguins.boxplot(column='flipper_length_mm', by='species', ax=axes[1, 0])
axes[1, 0].set_title('박스플롯: 종별 날개 길이')
axes[1, 0].set_xlabel('종')
axes[1, 0].set_ylabel('날개 길이 (mm)')

species_mean = penguins.groupby('species')['body_mass_g'].mean()
species_mean.plot(kind='bar', ax=axes[1, 1], color=['#EF5350', '#66BB6A', '#42A5F5'])
axes[1, 1].set_title('막대 그래프: 종별 평균 체중')
axes[1, 1].set_xlabel('종')
axes[1, 1].set_ylabel('평균 체중 (g)')
axes[1, 1].tick_params(axis='x', rotation=0)

fig.suptitle('Penguins 데이터셋 종합 실습', fontsize=15, fontweight='bold')
fig.tight_layout()
plt.show()
```
