## 6. 사용자 정의 함수(Custom Function) 생성

### minus 함수 설계

- **함수명:** `minus(data1, data2)`
- **반환값:** `return data1 - data2`
- **동작:** 두 입력값 간의 차이(Minus) 연산 수행
- `data1`: Series 개별 원소가 자동 할당
- `data2`: 추가 전달 인자

````python
# 사용자 정의 함수: minus
def minus(data1, data2):
    """두 입력값 간의 차이를 반환"""
    return data1 - data2

# 기본 호출 테스트
print(f"minus(50, 10) = {minus(50, 10)}")
print(f"minus(100, 30) = {minus(100, 30)}")
````

## 8. Lambda 함수 매핑

### Lambda 함수의 특징
- 별도 `def` 기반 사용자 함수 정의 과정 **생략**
- 코드 **간결화** 및 메모리 효율성 증대
- 일회성 속성 **실시간 연산에 최적화**

### 기본 문법
```python
df['fare'].apply(lambda x: minus(x, 10))
```
- 변수 `x`: Series의 개별 원소가 자동 할당되는 공간
- 익명 함수 구조를 통한 인자(x, 10) 직접 전달

### 8.1 Lambda 기반 매핑 코드 분석

| 기존 방식 (Standard) | Lambda 방식 (Anonymous) |
|---------------------|------------------------|
| `apply(minus, data2=10)` | `apply(lambda x: minus(x, 10))` |
| def 함수 정의 필요 | 인라인 정의 가능 |

````python
# 기존 방식 vs Lambda 방식 비교
print("=" * 60)
print("기존 방식 vs Lambda 방식 비교")
print("=" * 60)

# 방식 1: 기존 apply (함수명 + 추가 인자)
result_standard = df['fare'].apply(minus, data2=10)

# 방식 2: Lambda 방식
result_lambda = df['fare'].apply(lambda x: minus(x, 10))

print("\n[기존 방식] apply(minus, data2=10):")
print(result_standard.head())

print("\n[Lambda 방식] apply(lambda x: minus(x, 10)):")
print(result_lambda.head())

# 결과 동일성 검증
print(f"\n두 결과가 동일한가? {result_standard.equals(result_lambda)}")
````

````python
# Lambda만으로 함수 정의 없이 직접 연산
print("=" * 60)
print("Lambda만으로 직접 연산 (def 함수 정의 불필요)")
print("=" * 60)

# def 없이 Lambda만으로 동일한 연산
result_pure_lambda = df['fare'].apply(lambda x: x - 10)
print(result_pure_lambda.head())

# 결과 동일성 검증
print(f"\nminus 함수 결과와 동일한가? {result_standard.equals(result_pure_lambda)}")
````

## 9. DataFrame 전체 원소 대상 함수 매핑 분석

### 9.1 `applymap()` 함수 개요

- `applymap()` 함수 적용 원리
- 2차원 형태 데이터의 **전체 원소에 동시 접근**
- 기본 문법: `DataFrame객체.applymap(매핑 함수)`

### 9.2 `applymap()` 인자 전달 메커니즘

- DataFrame을 구성하는 **모든 개별 원소를 함수에 순차 입력**
- 입력된 2D 형태(Shape)를 **완벽히 유지**
- 입력된 구조의 **새로운 DataFrame 객체로 반환**

### 9.3 DataFrame 2D 연산을 위한 데이터 전처리

- 모든 열을 동시 연산을 위해 **수치형(Numeric) 데이터만 추출**
- 대상: 나이(age), 요금(fare)
- 단일 인자만 전달: `def minus(data): return data - 10`

````python
# DataFrame 2D 연산을 위한 데이터 전처리
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 수치형 데이터만 추출
df = titanic.loc[:, ['age', 'fare']]

print("=" * 60)
print("수치형 데이터 추출 (applymap용)")
print("=" * 60)
print(f"Shape: {df.shape}")
print(f"dtypes:\n{df.dtypes}")
print(f"\n상위 5행:")
print(df.head())
````

````python
# applymap용 단일 인자 함수 정의
# (applymap은 추가 인자 전달 불가 → 단일 인자 함수 사용)
def minus10(data):
    """입력값에서 10을 빼서 반환 (단일 인자)"""
    return data - 10

print(f"minus10(50) = {minus10(50)}")
print(f"minus10(7.25) = {minus10(7.25)}")
````

### 9.4 전체 원소 매핑 실행: `applymap()`

- 실행 코드: `df.applymap(minus10)`
- age (나이) 컬럼 전체 원소: -10 연산 적용
- fare (요금) 컬럼 전체 원소: -10 연산 동시 적용
- 다중 컬럼 **일괄 동시 적용**

````python
# applymap() 실행
# 참고: Pandas 2.1+ 에서는 applymap()이 map()으로 변경됨
# 호환성을 위해 두 가지 방법 모두 시도

print("=" * 60)
print("df.applymap(minus10) 실행 결과")
print("=" * 60)

try:
    # Pandas 2.1 이상
    result_map = df.map(minus10)
    print("(Pandas 2.1+: df.map() 사용)")
except AttributeError:
    # Pandas 2.0 이하
    result_map = df.applymap(minus10)
    print("(Pandas 2.0 이하: df.applymap() 사용)")

print(result_map)
````

### 9.5 `applymap()` 연산 결과 및 특성 분석

- age, fare **두 수치형 컬럼 모두 -10** 변환 확인
- 인덱스 0 결과: age(12.0), fare(-2.7500)
- 결측치 보존: 인덱스 888의 NaN(Not a Number)은 연산 후에도 **NaN으로 안전하게 유지**됨

````python
# 결과 분석
print("=" * 60)
print("applymap() 연산 결과 분석")
print("=" * 60)

print(f"\n결과 Shape: {result_map.shape} (원본과 동일)")
print(f"결과 dtypes:\n{result_map.dtypes}")

# 원본 vs 결과 비교 (상위 5행)
print("\n[원본]")
print(df.head())
print("\n[minus10 적용 결과]")
print(result_map.head())

# 결측치 보존 확인
print(f"\n결측치 보존 확인:")
print(f"  원본 age 결측: {df['age'].isnull().sum()}개")
print(f"  결과 age 결측: {result_map['age'].isnull().sum()}개")
print(f"  → NaN은 연산 후에도 NaN으로 안전하게 유지됨")
````

````python
# 결측치가 있는 행 확인
print("=" * 60)
print("결측치가 있는 행 비교 (원본 vs 결과)")
print("=" * 60)
nan_rows = df[df['age'].isnull()].head()
print("\n[원본 - NaN 행]:")
print(nan_rows)
print("\n[결과 - NaN 행]:")
print(result_map.loc[nan_rows.index])
print("\n→ age는 NaN 유지, fare는 정상적으로 -10 적용됨")
````

### 9.6 Lambda 연산 및 핵심 요약: `applymap()` + Lambda 매핑 비교

```python
df.applymap(lambda x: minus(x, 10))
```

- `applymap()`에도 Lambda 함수 적용 가능
- 단, `applymap()`에서는 추가 인자 전달이 직접 불가하므로 **Lambda를 통해 우회 전달**

````python
# applymap + Lambda 조합
print("=" * 60)
print("applymap + Lambda 조합")
print("=" * 60)

# 원본 minus 함수(2인자)를 Lambda로 감싸서 applymap에 적용
try:
    result_lambda_map = df.map(lambda x: minus(x, 10))
except AttributeError:
    result_lambda_map = df.applymap(lambda x: minus(x, 10))

print(result_lambda_map.head())

# 기존 결과와 동일한지 검증 (NaN 제외 비교)
print(f"\nminus10 함수 결과와 동일한가? {result_map.equals(result_lambda_map)}")
````

````python
# Lambda만으로 다양한 applymap 연산
print("=" * 60)
print("Lambda만으로 다양한 applymap 연산")
print("=" * 60)

# 제곱
try:
    print("\n[1] 제곱 연산:")
    print(df.map(lambda x: x ** 2).head())

    print("\n[2] 소수점 1자리 반올림:")
    print(df.map(lambda x: round(x, 1) if pd.notna(x) else x).head())

    print("\n[3] 2배 변환:")
    print(df.map(lambda x: x * 2).head())
except AttributeError:
    print("\n[1] 제곱 연산:")
    print(df.applymap(lambda x: x ** 2).head())

    print("\n[2] 소수점 1자리 반올림:")
    print(df.applymap(lambda x: round(x, 1) if pd.notna(x) else x).head())

    print("\n[3] 2배 변환:")
    print(df.applymap(lambda x: x * 2).head())
````

## 10. `apply()` vs `applymap()`: Pandas 함수 매핑 비교

| 구분 | `apply()` | `applymap()` |
|------|-----------|-------------|
| **타겟 객체** | Series 중심 | DataFrame 전용 |
| **데이터 차원** | 1차원 호출 제어 | 2차원 행렬 동시 제어 |
| **사용 목적** | 단일 속성 정밀 가공 | 다중 속성 일괄 스케일링/변환 |
| **추가 인자 전달** | `apply(func, arg=val)` 직접 전달 가능 | Lambda 우회 전달 필요 |

> **참고:** Pandas 2.1 이상에서는 `applymap()`이 `map()`으로 대체되었습니다.

````python
# apply() vs applymap() 종합 비교 실습
print("=" * 70)
print("apply() vs applymap() 종합 비교")
print("=" * 70)

df_demo = titanic.loc[:, ['age', 'fare']]

# apply(): Series 단위 - 단일 열에 적용
print("\n[apply()] Series 단위: fare 열에만 minus(data2=10) 적용")
result_apply = df_demo['fare'].apply(minus, data2=10)
print(result_apply.head())

# applymap(): DataFrame 전체 - 모든 열에 동시 적용
print("\n[applymap()] DataFrame 전체: age, fare 모두에 minus10 적용")
try:
    result_applymap = df_demo.map(minus10)
except AttributeError:
    result_applymap = df_demo.applymap(minus10)
print(result_applymap.head())
````

## 11. 종합 실습: Titanic 데이터 함수 매핑 파이프라인

````python
# 종합 실습 데이터 준비
df_full = titanic.loc[:, ['survived', 'sex', 'age', 'fare']].copy()
print("종합 실습 데이터:")
print(df_full.head(10))
````

````python
# 실습 1: apply()로 나이 구간 분류
def age_group(age):
    """나이를 구간별로 분류"""
    if pd.isna(age):
        return '미상'
    elif age < 18:
        return '미성년'
    elif age < 40:
        return '청장년'
    elif age < 60:
        return '중년'
    else:
        return '노년'

df_full['나이구간'] = df_full['age'].apply(age_group)
print("나이 구간 분류 결과:")
print(df_full['나이구간'].value_counts())
print()
print(df_full.head(10))
````

````python
# 실습 2: apply()로 요금 등급 분류 (Lambda 활용)
df_full['요금등급'] = df_full['fare'].apply(
    lambda x: '1등급' if x >= 50 else ('2등급' if x >= 20 else '3등급')
)
print("요금 등급 분류 결과:")
print(df_full['요금등급'].value_counts())
print()
print(df_full.head(10))
````

````python
# 실습 3: apply()로 성별 한글 변환
df_full['성별'] = df_full['sex'].apply(lambda x: '남성' if x == 'male' else '여성')
print("성별 한글 변환 결과:")
print(df_full[['sex', '성별']].head())
````

````python
# 실습 4: applymap()으로 수치형 데이터 정규화(0~1 스케일링)
df_numeric = df_full[['age', 'fare']].copy()

# 각 열의 min, max 확인
print("정규화 전:")
print(df_numeric.describe().loc[['min', 'max']])

# Min-Max 정규화 함수
def normalize(col):
    """Series를 0~1 범위로 정규화"""
    return (col - col.min()) / (col.max() - col.min())

# apply()로 열 단위 정규화 (axis=0)
df_normalized = df_numeric.apply(normalize)
print("\n정규화 후:")
print(df_normalized.describe().loc[['min', 'max']])
print()
print(df_normalized.head())
````

````python
# 실습 5: 나이구간별 생존율 분석 (파이프라인 종합)
print("=" * 60)
print("나이구간별 생존율 분석 (함수 매핑 파이프라인 종합)")
print("=" * 60)

survival_by_age = df_full.groupby('나이구간').agg(
    총_승객=('survived', 'count'),
    생존자=('survived', 'sum'),
    생존율=('survived', 'mean'),
    평균_요금=('fare', 'mean')
)
survival_by_age['생존율'] = (survival_by_age['생존율'] * 100).round(1).astype(str) + '%'
survival_by_age['평균_요금'] = survival_by_age['평균_요금'].round(2)
print(survival_by_age)
````
