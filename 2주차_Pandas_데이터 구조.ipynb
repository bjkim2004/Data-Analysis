## 0. 라이브러리 임포트

```python
import pandas as pd
import numpy as np

print(f'Pandas 버전: {pd.__version__}')
print(f'NumPy 버전: {np.__version__}')
```

# 1. Series 데이터

### 1.1 Series 생성: 리스트(List) 활용

```python
# 리스트로 Series 생성 (기본 인덱스: 0부터 시작)
data_list = [10, 25, 40, 55, 70]
ser1 = pd.Series(data_list)
print('=== 리스트 기반 Series ===')
print(ser1)
print(f'\ndtype: {ser1.dtype}')
print(f'index: {ser1.index}')
```

```python
# 사용자 정의 인덱스 지정
ser2 = pd.Series(data_list, index=['a', 'b', 'c', 'd', 'e'])
print('=== 사용자 정의 인덱스 Series ===')
print(ser2)
print(f'\nindex: {ser2.index}')
```

### 1.2 Series 생성: 딕셔너리(Dictionary) 활용

**핵심 포인트:** 딕셔너리의 Key값 → Series의 Index로 자동 변환 적용

```python
# 딕셔너리로 Series 생성
dic = {'name': ['hong', 'kim', 'heo'], 'kor': [80, 90, 75], 'eng': [80, 95, 100]}
ser_data = pd.Series(dic)
print('=== 딕셔너리 기반 Series ===')
print(ser_data)
print(f'\ndtype: {ser_data.dtype}')
```

```python
# 딕셔너리의 Key가 Index로 변환됨을 확인
print('=== Index 확인 ===')
print(f'Index: {ser_data.index}')
print(f'Index 리스트: {ser_data.index.tolist()}')
```

### 1.3 Series 생성: NumPy Array 활용

**핵심 포인트:** 인덱스 미지정 시 → 0-based index 자동 할당

```python
# NumPy 배열로 Series 생성
arr = np.array([80, 90, 75])
ser_data = pd.Series(arr)
print('=== NumPy 배열 기반 Series ===')
print(ser_data)
print(f'\ndtype: {ser_data.dtype}')
print(f'index: {ser_data.index}  # 0-based index 자동 할당')
```

### 1.4 Series의 주요 특징: 인덱싱

```python
# 정수 기반 인덱싱
ser = pd.Series([100, 200, 300, 400, 500])
print('=== 정수 기반 인덱싱 ===')
print(f'ser[0] = {ser[0]}')
print(f'ser[2] = {ser[2]}')
print(f'ser[1:4] =\n{ser[1:4]}')
```

```python
# 레이블 기반 인덱싱
ser_label = pd.Series([100, 200, 300], index=['apple', 'banana', 'cherry'])
print('=== 레이블 기반 인덱싱 ===')
print(f"ser_label['apple'] = {ser_label['apple']}")
print(f"ser_label['cherry'] = {ser_label['cherry']}")
```

### 1.5 Series의 주요 특징: 연산 및 누락 데이터

```python
# NaN이 포함된 Series의 통계 메서드
ser_nan = pd.Series([15, 30, 45, np.nan])
print('=== NaN 포함 Series ===')
print(ser_nan)
print(f'\nsum()  = {ser_nan.sum()}    # NaN 자동 제외')
print(f'mean() = {ser_nan.mean()}   # NaN 자동 제외')
print(f'count()= {ser_nan.count()}     # NaN 제외한 유효 데이터 수')
```

```python
# 서로 다른 길이의 Series 간 연산
ser_a = pd.Series([10, 20, 30], index=['x', 'y', 'z'])
ser_b = pd.Series([1, 2, 3, 4], index=['x', 'y', 'z', 'w'])

print('=== 서로 다른 길이의 Series 연산 ===')
print('ser_a + ser_b =')
print(ser_a + ser_b)
print('\n# 매칭되지 않는 인덱스(w)는 NaN 처리됨')
```

# 2. DataFrame

### 2.1 DataFrame 초기화: NumPy 배열(Array) 기반 생성

```python
# NumPy 배열 기반 DataFrame 생성
arr = np.array([80, 90, 75])
df = pd.DataFrame(arr)
print('=== 1D 배열 기반 DataFrame ===')
print(df)
print(f'\n행 인덱스: {df.index.tolist()}')
print(f'열 인덱스: {df.columns.tolist()}')
print('\n# np.array 데이터를 입력으로 사용함.')
print('# 행 인덱스(0,1,2)와 열 인덱스(0)가 엔진에 의해 자동 부여됨.')
```

```python
# 2D 배열 기반 DataFrame 생성
arr_2d = np.array([[80, 90], [75, 85], [90, 95]])
df_2d = pd.DataFrame(arr_2d, columns=['국어', '영어'], index=['hong', 'kim', 'heo'])
print('=== 2D 배열 + 열/행 레이블 ===')
print(df_2d)
```

### 2.2 DataFrame 초기화: 딕셔너리(Dictionary) 기반 생성

```python
# 딕셔너리 기반 DataFrame 생성
dic = {'name': ['hong', 'kim', 'heo'],
       'kor': [80, 90, 75],
       'eng': [80, 95, 100]}

df = pd.DataFrame(dic)
print('=== 딕셔너리 기반 다중 열 생성 ===')
print(df)
print(f'\n# dic 객체의 Key가 열 방향 메타데이터로 즉각 적용됨.')
print(f'# 구조화된 관계형 데이터셋 구축에 최적화된 방식임.')
```

### 2.3 DataFrame 확장성과 유연성

- **이기종 수용**: 서로 다른 데이터 타입의 항목 혼합 가능
- **가변성**: 데이터 크기(Size) 자유로운 변경
- **인덱스 중복**: 인덱스 중복 가능 및 동일 인덱스 내 다른 행 값 허용
- **연산 기능**: 행과 열 기준의 산술 연산 완벽 수행

```python
# 이기종 데이터 타입 혼합
df_mixed = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'score': [88.5, 92.3, 78.1],
    'pass': [True, True, False]
})
print('=== 이기종 데이터 타입 ===')
print(df_mixed)
print(f'\n각 열의 dtype:\n{df_mixed.dtypes}')
```

```python
# 인덱스 중복 허용 확인
df_dup = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Alice'],
    'age': [25, 30, 25],
    'city': ['NYC', 'LA', 'NYC']
}, index=['a', 'b', 'a'])  # 인덱스 'a'가 중복

print('=== 인덱스 중복 허용 ===')
print(df_dup)
print(f"\ndf_dup.loc['a'] =\n{df_dup.loc['a']}")
print('\n# 중복된 키값을 호출할 경우, 해당 값을 가진 모든 행(Row) 데이터를 일괄 반환함')
```

### 2.4 Series vs DataFrame 비교

| 항목 | Series | DataFrame |
|:---:|:---:|:---:|
| 차원 | 1차원 (1D) | 2차원 (2D) |
| 구성 | 단일 열 데이터 | 여러 Series의 집합 |
| 인덱스 | Row Index | Row Index & Column Labels |
| 데이터 타입 | 단일 타입 권장 | 열(Column)별 이기종 타입 허용 |

```python
# Series와 DataFrame 비교 실습
ser_example = pd.Series([80, 90, 75], index=['hong', 'kim', 'heo'], name='kor')
df_example = pd.DataFrame({'kor': [80, 90, 75], 'eng': [80, 95, 100]},
                           index=['hong', 'kim', 'heo'])

print('=== Series ===')
print(ser_example)
print(f'type: {type(ser_example)}')
print(f'ndim: {ser_example.ndim}')

print('\n=== DataFrame ===')
print(df_example)
print(f'type: {type(df_example)}')
print(f'ndim: {df_example.ndim}')
```

```python
# DataFrame의 각 열은 Series임을 확인
print('=== DataFrame의 열 = Series ===')
print(f"type(df_example['kor']) = {type(df_example['kor'])}")
print(f"\ndf_example['kor'] =\n{df_example['kor']}")
```

# 3. 색인(Index) 메커니즘

### 3.1 Index 객체의 코어 기능

```python
# DataFrame의 index와 columns 속성 확인
df = pd.DataFrame({'name': ['hong', 'kim', 'heo'],
                    'kor': [80, 90, 75],
                    'eng': [80, 95, 100]})

print('=== 기본 인덱스/컬럼 확인 ===')
print(f'df.index   = {df.index}')
print(f'df.columns = {df.columns}')
print(f'\nindex type  : {type(df.index)}')
print(f'columns type: {type(df.columns)}')
```

### 3.2 Index 아키텍처 특징 [1/3]: 불변성과 순서


```python
# Index의 불변성 확인
idx = pd.Index([1, 2, 3, 4, 5])
print(f'Index: {idx}')

# 개별 요소 변경 시도 → 에러 발생
try:
    idx[0] = 100
except TypeError as e:
    print(f'\nTypeError: {e}')
    print('# Index 객체는 불변(Immutable)이므로 개별 요소 변경 불가')
```

```python
# Index 슬라이싱
idx = pd.Index([10, 20, 30, 40, 50])
print(f'idx[1:4] = {idx[1:4]}')
print(f'idx[:3]  = {idx[:3]}')
```

### 3.3 Index 아키텍처 특징 [2/3]: 공유와 고정 스펙

- 객체를 미리 생성하여 여러 자료구조 간 안전한 공유(Sharing)가 가능함
- 생성된 색인 객체의 크기(Size)는 물리적으로 변경 불가능함
- 단, `reindex` 명령을 통한 새로운 레이아웃 매핑은 예외적으로 지원됨

```python
# Index 객체 공유 예시
shared_idx = pd.Index(['A', 'B', 'C'])

ser1 = pd.Series([10, 20, 30], index=shared_idx)
ser2 = pd.Series([100, 200, 300], index=shared_idx)

print('=== Index 객체 공유 ===')
print(f'ser1.index is ser2.index: {ser1.index is ser2.index}')
print(f'\nser1:\n{ser1}')
print(f'\nser2:\n{ser2}')
```

### 3.4 Index 아키텍처 특징 [3/3]: 다중 선택(Duplication) 매커니즘

```python
# 중복 인덱스 예시
ser_dup = pd.Series([10, 20, 30, 40], index=['A', 'B', 'A', 'C'])
print('=== 중복 인덱스 ===')
print(ser_dup)
print(f"\nser_dup['A'] =\n{ser_dup['A']}")
print('\n# 중복된 인덱스 A에 해당하는 모든 값이 반환됨')
```
