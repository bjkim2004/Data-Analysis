## Pandas의 인덱스(Index) 처리

## 1. 위치 기반 인덱싱 개념

```python
import pandas as pd

df = pd.DataFrame(
    data=[
        ['홍길동', 80, 90, 85],
        ['이기자', 70, 75, 85],
        ['최신', 100, 90, 95]
    ],
    index=[1, 2, 3],
    columns=['name', 'kor', 'eng', 'mat']
)
df
```

## 2. 기본 슬라이싱 코드

```python
df[0:1]
```

## 3. .iloc 단일 행 추출

```python
df.iloc[0]
```

## 4. .iloc 복합 추출 (행 + 열)

```python
df.iloc[0:2, [0, 2]]
```

## 5. 불리언 인덱싱 (Boolean Indexing)

### 5-1. 불리언 조건식 생성

```python
df.eng >= 90
```

### 5-2. 불리언 인덱싱 적용

```python
df[df.eng >= 90]
```

### 5-3. .loc와 불리언의 결합

```python
df.loc[df.eng >= 90, ['name', 'eng']]
```

## 6. Pandas 인덱싱 마스터 요약
- 기본 슬라이싱: df[]
- 위치 기반: df.iloc[]
- 라벨 기반: df.loc[]

## 7. 특정 열을 인덱스로 (set_index)

### 7-1. set_index 실행을 위한 데이터 준비

```python
import pandas as pd

df = pd.DataFrame(
    data=[
        ['홍길동', 80, 90, 85],
        ['이기자', 70, 75, 85],
        ['최신', 100, 90, 95]
    ],
    index=[1, 2, 3],
    columns=['name', 'kor', 'eng', 'mat']
)
df
```

### 7-2. 'name' 칼럼 인덱스 승격

```python
df.set_index('name')
```

### 7-3. set_index 파라미터 - keys, drop

```python
df.set_index('name', drop=True)
```

```python
df.set_index('name', drop=False)
```

### 7-4. set_index 파라미터 - append, inplace

```python
df.set_index('name', append=True)
```

```python
df.set_index('name', inplace=True)
df
```

## 8. 인덱스 초기화 (reset_index)

### 8-1. 인덱스 완전 삭제 및 초기화 (drop=True)

```python
df.reset_index(drop=True)
```

### 8-2. 보존형 인덱스 초기화 (Default, drop=False)

```python
df.reset_index()
```
