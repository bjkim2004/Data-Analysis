## Step 2. 행과 열 추가/삭제하기

### 행(Row) 추가: .loc[] 인덱서

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
print("Before:")
print(df)

df.loc[3] = ['song', 95, 80]
print("\nAfter:")
print(df)
```

### 단일 열 추가 아키텍처

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
print("Before:")
print(df)

df['math'] = [70, 85, 90]
print("\nAfter:")
print(df)
```

### 행/열 삭제: drop() 메서드

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
df
```

### axis=0 (행 삭제) vs axis=1 (열 삭제)

```python
# axis=0: 행 삭제
df_row_dropped = df.drop(0, axis=0)
print("행 삭제 (index=0):")
print(df_row_dropped)
```

```python
# axis=1: 열 삭제
df_col_dropped = df.drop('eng', axis=1)
print("열 삭제 (eng):")
print(df_col_dropped)
```

### inplace 파라미터

```python
# inplace=False (기본값): 원본 보존, 새로운 DataFrame 반환
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
new_df = df.drop(0, axis=0)
print("원본 df:")
print(df)
print("\nnew_df:")
print(new_df)
```

```python
# inplace=True: 원본 DataFrame 자체를 즉시 변형
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
df.drop(0, axis=0, inplace=True)
print("원본 df (inplace=True 적용 후):")
print(df)
```

## Step 3. 행, 열, 원소 선택하기

### 행 추출 1: loc 메커니즘

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})

# 단일 행 선택
print("df.loc[0]:")
print(df.loc[0])
```

```python
# 범위 지정 선택 (끝행 포함)
print("df.loc[0:1]:")
print(df.loc[0:1])
```

### 행 추출 2: iloc 메커니즘

```python
# 단일 행 선택
print("df.iloc[2]:")
print(df.iloc[2])
```

```python
# 범위 지정 및 슬라이싱 (끝인덱스 제외)
print("df.iloc[1:3]:")
print(df.iloc[1:3])
```

### iloc 슬라이싱 심화 규칙

```python
# 역방향 추출 (Step = -1)
print("df.iloc[::-1]:")
print(df.iloc[::-1])
```

```python
# 전체 추출
print("df.iloc[:]:")
print(df.iloc[:])
```

### 열(Column) 타겟팅: 단일 vs 다중

```python
# [1차원] 1개의 열 선택 -> Series
print("df['name']:")
print(df['name'])
print(type(df['name']))
```

```python
# [2차원] 여러 개의 열 선택 -> DataFrame
print("df[['name','kor']]:")
print(df[['name','kor']])
print(type(df[['name','kor']]))
```

### 원소 정밀 조준 1: loc 교차점

```python
# 단일 추출: hong의 eng 점수
print("df.loc[0, 'eng']:")
print(df.loc[0, 'eng'])
```

```python
# 다중 추출: kim의 eng, kor 점수
print("df.loc[1, ['eng', 'kor']]:")
print(df.loc[1, ['eng', 'kor']])
```

### 원소 정밀 조준 2: iloc 좌표계

```python
# 단일 추출: hong의 eng 점수 (행 0, 열 2)
print("df.iloc[0, 2]:")
print(df.iloc[0, 2])
```

```python
# 다중 추출: kim의 eng, kor 점수 (행 1, 열 [2, 1])
print("df.iloc[1, [2, 1]]:")
print(df.iloc[1, [2, 1]])
```

## Step 4. 데이터 위치 변경하기

### 행 위치 재조합: reindex()

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
print("Before:")
print(df)

df = df.reindex(index=[2, 0, 1])
print("\nAfter (reindex=[2,0,1]):")
print(df)
```

### 열 위치 재조합: 대괄호 재나열

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'eng':[80,95,100],'kor':[80,90,75]})
print("Before:")
print(df)

df = df[['name', 'kor', 'eng']]
print("\nAfter (열 순서 변경):")
print(df)
```

### 축 완전 변환: 행/열 전치 (Transpose)

```python
df = pd.DataFrame({'name':['hong','kim','heo'],'kor':[80,90,75],'eng':[80,95,100]})
print("원본:")
print(df)

# 정식 메서드
print("\ndf.transpose():")
print(df.transpose())

# 단축 속성
print("\ndf.T:")
print(df.T)
```
